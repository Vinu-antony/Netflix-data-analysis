```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
df=pd.read_csv("netflix_titles.csv")
df.head(10)
df.info()
df.describe()
df.isnull().sum()
# Drop rows with missing values
df.dropna(subset=['director','cast'],inplace=True)
df.fillna({'rating':'Not rated'},inplace=True)
df = df.dropna(subset=['date_added'])
invalid_dates = df[~df['date_added'].str.match(r'\w+ \d{1,2}, \d{4}', na=False)]
print(invalid_dates)
sns.countplot(x=df['type'], palette='coolwarm')
plt.title('Distribution of Movies and TV Shows on Netflix')
plt.show()
top_countries = df['country'].value_counts().head(10)
top_countries.plot(kind='bar', color='teal')
plt.title('Top 10 Countries with Most Netflix Titles')
plt.xlabel('Country')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()
from collections import Counter

genres = ','.join(df['listed_in'].dropna()).split(', ')
genre_counts = Counter(genres)
top_genres = genre_counts.most_common(10)

# Convert to DataFrame for visualization
genre_df = pd.DataFrame(top_genres, columns=['Genre', 'Count'])
sns.barplot(x='Count', y='Genre', data=genre_df, palette='magma')
plt.title('Top 10 Most Popular Genres on Netflix')
plt.show()

df['release_year'].value_counts().sort_index().plot(kind='line', marker='o', color='green')
plt.title('Number of Releases Per Year')
plt.xlabel('Year')
plt.ylabel('Count')
plt.show()

from wordcloud import WordCloud

text = ' '.join(df['title'].dropna())
wordcloud = WordCloud(width=800, height=400, background_color='black').generate(text)

plt.figure(figsize=(10,5))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.show()












```
